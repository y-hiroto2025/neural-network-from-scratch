```python
import numpy as np

def softmax(x):
  """
  前層からの出力を確率に変換する
  """
  c = np.max(x)
  exp_a = np.exp(x - c)     # オーバーフロー対策

  sum_exp_a = np.sum(exp_a)

  output = exp_a / sum_exp_a
  return output
```

```python
print(softmax(np.array([2.5, 0.3, 1.2])))
```
